0. 导入库

```python
import numpy as npy
import pandas as pds
from sklearn.metrics import classification_report
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import AdaBoostClassifier
from sklearn import datasets
import matplotlib.pyplot as plt
```

1\. 装载数据集并且显示一个样板

```python
# load the MNIST digits dataset
mnist = datasets.load_digits()

X = npy.array(mnist.data)
Y = npy.array(mnist.target)
n = len(X)

def display(indexes, labels):
    for i in range(len(indexes))[:24]:
        plot = plt.subplot(4, 6, i + 1)
        plot.text(-2, 0, labels[i], bbox={'facecolor': 'white', 'pad': 2})
        plt.imshow(mnist.images[indexes[i]])

display(range(n), Y)
```

2\. 切割训练集、测试集（固定比例 $7:3$）

```python
train = npy.random.choice(n, round(0.7 * n))
test = [i for i in range(n) if i not in train]
```

3\. 用 scikit learn 的 DecisionTreeClassifier 进行分类模型训练

```python
dtc = DecisionTreeClassifier()
dtc.fit(X[train], Y[train])
```

4\. 测试集上的分类效果，即每个类别的 precision, recall, f1

```python
def show_classification_report(Y, predict):
    cr = classification_report(Y, predict, output_dict=True)
    return pds.DataFrame(cr).transpose()

train_predict_dtc = dtc.predict(X[train])
show_classification_report(Y[train], train_predict_dtc)
```

```python
test_predict_dtc = dtc.predict(X[test])
show_classification_report(Y[test], test_predict_dtc)
```

5\. 用 AdaBoostClassifier 包装 DecisionTreeClassifier，完成分类模型训练

```python
abc = AdaBoostClassifier(base_estimator=dtc)
abc.fit(X[train], Y[train])
```

6\. 测试集上的分类效果，即每个类别的 precision, recall, f1

```python
train_predict_abc = abc.predict(X[train])
show_classification_report(Y[train], train_predict_abc)
```

```python
test_predict_abc = abc.predict(X[test])
show_classification_report(Y[test], test_predict_abc)
```
