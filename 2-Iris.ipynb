# 2.2 插入一列，名称为color
1,根据species的值，设定color列的值如下

Iris-setosa  --> 0

Iris-versicolor -->1

Iris-virginica -->2

2，查看首部5行

3，查看50-55行

4，查看尾部5行


```python
new_iris_data = iris_data.copy()
value_map = {
    "Iris-setosa": 0,
    "Iris-versicolor": 1,
    "Iris-virginica": 2
}
new_iris_data.insert(column="color", value=[value_map[species[i]] for i in range(len(species))], loc=len(iris_data.columns))
new_iris_data
```

# 把所有的样本点用scatter显示
1,横坐标，从[sepal length,sepal width,petal length,petal width]选择一个属性

2,纵坐标，从[sepal length,sepal width,petal length,petal width]选择另一个属性

提示，也就是横坐标和纵坐标是[sepal length,sepal width,petal length,petal width]的两两组合，比如sepal length-sepal length，但是不要跟自己组合，比如sepal length-sepal length

3,散点颜色为1、2、3，分别表示Iris-setosa、Iris-versicolor、Iris-virginica

4,看看各个类别时间是否有比较清晰的分类边界，考察两个属性是否可以把各个类别分开
